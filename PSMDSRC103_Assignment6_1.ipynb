```python
#Median without import package
#Since the length of the list is 100(Even) I need to get the average of the 2 middle values.
def median(list):
    sorted_list = sorted(list)
    count = len(sorted_list)
    m1 = sorted_list[count // 2 - 1]
    m2 = sorted_list[count // 2]
    return (m1 + m2) / 2

median(salaries)
```

```python
#Median with import package
import statistics as st
median = st.median(salaries)
print(median)
```

```python
#Mode without import package
def mode(list):
    frequency = {}
    max_count = 0
    most_frequent_item = None
    
    for item in list:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1
        
        if frequency[item] > max_count:
            max_count = frequency[item]
            most_frequent_item = item
    
    return most_frequent_item

mode(salaries)
```
